```python
import mysql.connector


class EmployeeManagement:
    def __init__(self):
        self.db = mysql.connector.connect(
            host="localhost",
            user="root",
            password="1234",
            database="EmployeeData"
        )
        self.cursor = self.db.cursor()
        self.create_table()

    def create_table(self):
        self.cursor.execute("""CREATE TABLE IF NOT EXISTS EMPLOYEE(
                id INT AUTO_INCREMENT PRIMARY KEY,
                name VARCHAR(255),
                salary DECIMAL(10,2),
                age INT,
                position VARCHAR(255) )
        """)
        self.db.commit()

    def add_e(self, name, salary, age, position):
        sql = "INSERT INTO EMPLOYEE(name,salary,age,position) VALUES(%s,%s,%s,%s)"
        val = (name, salary, age, position)
        self.cursor.execute(sql, val)
        self.db.commit()
        print(f"New Employee with {name} are Added Successfully")

    def remove_e_by_id(self, id):
        sql = "DELETE FROM EMPLOYEE WHERE id=%s"
        val = (id,)
        self.cursor.execute(sql, val)
        self.db.commit()
        print(f"Employee with ID {id} is deleted successfully")

    def remove_e_by_name(self, name):
        sql = "DELETE FROM EMPLOYEE WHERE name=%s"
        val = (name,)
        self.cursor.execute(sql, val)
        self.db.commit()
        print(f"Employee with Name {name} is deleted successfully")

    def update_e_by_id(self, id, name, salary, age, position):
        sql = "UPDATE EMPLOYEE SET name=%s ,salary=%s, age=%s, position=%s WHERE id=%s"
        val = (name, salary, age, position, id,)
        self.cursor.execute(sql, val)
        self.db.commit()
        print(f"Employee with ID {id} are updated successfully")

    def update_e_by_name(self, old_name, new_name, salary, age, position):
        sql = "UPDATE EMPLOYEE SET new_name=%s, salary=%s, age=%s, position=%s where old_name=%s"
        val = (old_name, new_name, salary, age, position,)
        self.cursor.execute(sql, val)
        self.db.commit()
        print(f"Employee with Name {old_name} are updated successfully")

    def display_e(self):
        self.cursor.execute("SELECT * FROM EMPLOYEE")
        result = self.cursor.fetchall()
        for x in result:
            print(f"ID:{x[0]}, Name:{x[1]}, Salary:{x[2]}, Age:{x[3]}, Position:{x[4]}")
        #     print("Employee ID:- "+str(x[0])+" Employee Name:- "+str(x[1]),"Employee Salary:- "
        #     +str(x[2]), "Employee Age:-" +str(x[3]), "Employee Position:-" +str(x[4]))

    def close(self):
        self.cursor.close()
        self.db.close()

    # def load(self):
    #     print("Data is Loaded")
    #     self.display_e()


e1 = EmployeeManagement()
while True:
    print("\n Employee Management System Menu")
    print("1. Add Employees")
    print("2. Remove Employee By ID")
    print("3. Remove Employee By Name")
    print("4. Update Employee By ID")
    print("5. Update Employee By Name")
    print("6. Display All Employees")
    print("7. Exit")
    choice = int(input("Enter the Choice:"))
    if choice == 1:
        name = input("Enter the Name:")
        salary = float(input("Enter the Salary:"))
        age = int(input("Enter the Age:"))
        position = input("Enter the Position:")
        e1.add_e(name, salary, age, position)
    elif choice == 2:
        id = int(input("Enter the ID to be Removed:"))
        e1.remove_e_by_id(id)
    elif choice == 3:
        name = input("Enter the Name to be Removed:")
        e1.remove_e_by_name(name)
    elif choice == 4:
        id = int(input("Enter the Id to be Updated:"))
        name = input("Enter the Name (Press Enter to Skip):")
        salary = input("Enter the Salary (Press Enter to Skip):")
        salary = float(salary) if salary else None
        age = input("Enter the Age(Press Enter to Skip):")
        age = int(age) if age else None
        position = input("Enter the Position(Press Enter to Skip):")
        e1.update_e_by_id(id, name, salary, age, position)
    elif choice == 5:
        old_name = input("Enter the Current Name of Employee to be Updated:")
        new_name = input("Enter the New Name(Press Enter to Skip):")
        salary = input("Enter the Salary (Press Enter to Skip) :")
        salary = float(salary) if salary else None
        age = input("Enter the Age(Press Enter to Skip) :")
        age = int(age) if age else None
        position = input("Enter the Position(Press Enter to Skip):")
        e1.update_e_by_name(old_name, new_name, salary, age, position)
    elif choice == 6:
        e1.display_e()
    elif choice == 7:
        print("Exit the Program")
        e1.close()
        break
    else:
        print("Wrong Choice- Enter the Valid Choice")
```
